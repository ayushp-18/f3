```python
# Cell 4: Parser helpers (we will tweak these quickly)
def clean_num(s):
    if not s: return None
    s = str(s).strip().replace(',', '').replace('','').replace('$','')
    if s in ['', '-', '--']: return None
    m = re.findall(r'[-+]?\d*\.\d+|\d+', s.replace(' ', ''))
    return float(m[0]) if m else None

# regexes
numeric_suffix_re = re.compile(r"""
    (?P<prefix>.*\S)\s+
    (?P<qty>\d+(?:\.\d+)?)\s+
    (?P<rate>[-\d,\.]+)\s+
    (?P<discount>[-\d,\.]+)\s+
    (?P<net>[-\d,\.]+)\s*$
    """, re.VERBOSE)

amount_only_re = re.compile(r'(?P<prefix>.*\S)\s+(?P<net>[-\d,\.]+)\s*$')

def parse_page_text(text):
    lines = [ln.strip() for ln in text.splitlines() if ln.strip()!='']
    items = []
    buf_name = None
    for ln in lines:
        m = numeric_suffix_re.match(ln)
        if m:
            name = m.group('prefix').strip()
            qty = clean_num(m.group('qty'))
            rate = clean_num(m.group('rate'))
            net = clean_num(m.group('net'))
            items.append({
                'item_name': name,
                'item_quantity': qty if qty is not None else 1.0,
                'item_rate': rate,
                'item_amount': net
            })
            buf_name = None
            continue
        m2 = amount_only_re.match(ln)
        if m2:
            possible_amount = clean_num(m2.group('net'))
            if possible_amount is not None and buf_name is not None:
                name = (buf_name + ' ' + m2.group('prefix')).strip()
                items.append({
                    'item_name': name,
                    'item_quantity': 1.0,
                    'item_rate': None,
                    'item_amount': possible_amount
                })
                buf_name = None
                continue
        low = ln.lower()
        if any(h in low for h in ['description', 'qty', 'rate', 'discount', 'net amt', 'total']):
            buf_name = None
            continue
        buf_name = (buf_name + ' ' + ln) if buf_name else ln
    return items
```

```python
# Cell 7  improved filtering for valid bill pages

VALID_KEYWORDS = [
    "consultation", "charge", "room", "rent", "nursing",
    "blood", "test", "fee", "scan", "surgeon", "ot", "gas"
]

def is_junk_page(text):
    # if OCR mistakenly reads JSON itself
    if "item quantity" in text.lower():
        return True
    if "item name" in text.lower():
        return True
    if "pagewise line items" in text.lower():
        return True
    return False

def is_valid_bill_page(items, text):
    # Rule 1: at least 3 line items
    if len(items) < 3:
        return False
    # Rule 2: at least 1 bill-related keyword
    low = text.lower()
    if not any(k in low for k in VALID_KEYWORDS):
        return False
    # Rule 3: should not be junk JSON
    if is_junk_page(text):
        return False
    return True

filtered_results = []

for i, page in enumerate(pages, start=1):
    txt = pytesseract.image_to_string(page)
    items = parse_page_text(txt)
    if is_valid_bill_page(items, txt):
        page_type = 'Final Bill' if re.search(r'\b(total|grand total|amount due)\b', txt, re.I) else 'Bill Detail'
        filtered_results.append({
            "page_no": str(i),
            "page_type": page_type,
            "bill_items": items
        })

clean_output = {
    "is_success": True,
    "token_usage": {"total_tokens": 0, "input_tokens": 0, "output_tokens": 0},
    "data": {
        "pagewise_line_items": filtered_results,
        "total_item_count": sum(len(p['bill_items']) for p in filtered_results)
    }
}

print(json.dumps(clean_output, indent=2, ensure_ascii=False))
```

```python
# Cell 8  stronger junk filtering

JUNK_PATTERNS = [
    "tem_amount", "tem quantity", "item name\":", "item amount\":",
    "item rate\":", "item quantity\":", "pagewise", "response format",
    "sr=", "sig=", "sp=", "se="
]

def strong_is_junk_page(text):
    low = text.lower()
    for pat in JUNK_PATTERNS:
        if pat in low:
            return True
    return False

def final_is_valid_bill_page(items, text):
    # At least 3 items AND contains medical/billing keywords AND not junk
    low = text.lower()

    if strong_is_junk_page(text):
        return False

    if len(items) < 3:
        return False

    if not any(k in low for k in VALID_KEYWORDS):
        return False

    return True

# Apply final filtering
final_results = []

for i, page in enumerate(pages, start=1):
    txt = pytesseract.image_to_string(page)
    items = parse_page_text(txt)

    if final_is_valid_bill_page(items, txt):
        page_type = 'Final Bill' if re.search(r'\b(total|grand total|amount due)\b', txt, re.I) else 'Bill Detail'
        final_results.append({
            "page_no": str(i),
            "page_type": page_type,
            "bill_items": items
        })

final_output = {
    "is_success": True,
    "token_usage": {"total_tokens": 0, "input_tokens": 0, "output_tokens": 0},
    "data": {
        "pagewise_line_items": final_results,
        "total_item_count": sum(len(p["bill_items"]) for p in final_results)
    }
}

print(json.dumps(final_output, indent=2, ensure_ascii=False))
```

```python
# Cell X: debug per-page and auto-select best pages
JUNK_PATTERNS = [
    "tem_amount", "tem quantity", "item name\":", "item amount\":",
    "item rate\":", "item quantity\":", "pagewise", "response format",
    "sr=", "sig=", "sp=", "se="
]
VALID_KEYWORDS = [
    "consultation", "charge", "room", "rent", "nursing",
    "blood", "test", "fee", "scan", "surgeon", "ot", "gas"
]

def count_keyword_matches(text):
    low = text.lower()
    return sum(1 for k in VALID_KEYWORDS if k in low)

def count_junk_matches(text):
    low = text.lower()
    return sum(1 for p in JUNK_PATTERNS if p in low)

page_infos = []
for i, page in enumerate(pages, start=1):
    txt = pytesseract.image_to_string(page)
    items = parse_page_text(txt)
    items_count = len(items)
    kw_matches = count_keyword_matches(txt)
    junk_matches = count_junk_matches(txt)
    score = items_count + 2*kw_matches - 10*junk_matches
    snippet = txt.replace('\n',' ')[:400]
    page_infos.append({
        "page_no": i,
        "items_count": items_count,
        "keyword_matches": kw_matches,
        "junk_matches": junk_matches,
        "score": score,
        "snippet": snippet
    })

# Print diagnostics
print("Per-page diagnostics:")
for info in page_infos:
    print(f"--- Page {info['page_no']} ---")
    print(f" items_count: {info['items_count']}, keyword_matches: {info['keyword_matches']}, junk_matches: {info['junk_matches']}, score: {info['score']}")
    print(" snippet:", info['snippet'][:300])
    print()

# Auto-select pages with score >= 1
selected = [p for p in page_infos if p['score'] >= 1]
print("Auto-selected pages (score >= 1):", [p['page_no'] for p in selected])

# Build final_results from selected pages
final_results = []
for p in selected:
    i = p['page_no']
    txt = pytesseract.image_to_string(pages[i-1])
    items = parse_page_text(txt)
    page_type = 'Final Bill' if re.search(r'\b(total|grand total|amount due)\b', txt, re.I) else 'Bill Detail'
    final_results.append({
        "page_no": str(i),
        "page_type": page_type,
        "bill_items": items
    })

final_output = {
    "is_success": True,
    "token_usage": {"total_tokens": 0, "input_tokens": 0, "output_tokens": 0},
    "data": {
        "pagewise_line_items": final_results,
        "total_item_count": sum(len(p["bill_items"]) for p in final_results)
    }
}

print("\nFinal extracted JSON (from selected pages):")
print(json.dumps(final_output, indent=2, ensure_ascii=False))
```

```python
# Cell: build master list, dedupe, detect subtotals and totals

import re, json
from difflib import SequenceMatcher
from collections import OrderedDict

# helper functions
def normalize_name(s):
    s = re.sub(r'[^A-Za-z0-9 ]+', ' ', s).lower().strip()
    s = re.sub(r'\s+', ' ', s)
    return s

def similar(a,b):
    a1 = normalize_name(a)
    b1 = normalize_name(b)
    return SequenceMatcher(None, a1, b1).ratio()

# collect all extracted items from final selection (use final_results from previous cell)
# if final_results not present, try 'final_results' or 'filtered_results' or parsed results
candidates = []
try:
    source_pages = final_results
except NameError:
    try:
        source_pages = filtered_results
    except NameError:
        source_pages = results  # fallback

for p in source_pages:
    for it in p['bill_items']:
        item = dict(it)  # copy
        item['_page_no'] = p['page_no']
        candidates.append(item)

# build master list with dedupe
master = []
for it in candidates:
    amt = it.get('item_amount')
    name = it.get('item_name','')
    qty = it.get('item_quantity',1.0)
    rate = it.get('item_rate')
    norm = normalize_name(name)
    keep = True
    for existing in master:
        # if amounts match exactly and names very similar -> duplicate
        if existing['item_amount'] is not None and amt is not None and abs(existing['item_amount'] - amt) < 0.01:
            sim = similar(existing['item_name'], name)
            if sim > 0.75:
                # prefer the one from lower page_no (already ordered), skip this
                keep = False
                break
        # if name is substring of other and amount equals -> duplicate
        if amt is not None and existing['item_amount'] is not None and abs(existing['item_amount'] - amt) < 0.01:
            if norm in normalize_name(existing['item_name']) or normalize_name(existing['item_name']) in norm:
                keep = False
                break
    if keep:
        master.append(it)

# compute sum of unique item amounts
sum_extracted = sum((it.get('item_amount') or 0.0) for it in master)

# detect subtotal-like lines in OCR pages (search page texts)
subtotal_candidates = []
final_total_candidates = []

all_page_texts = []
for i, page in enumerate(pages, start=1):
    txt = pytesseract.image_to_string(page)
    all_page_texts.append((i, txt))
    # find lines like "Room Charges 4,850.00" or "Room Charges 4850.00"
    for ln in txt.splitlines():
        ln = ln.strip()
        if not ln: continue
        m = re.search(r'([A-Za-z &/\(\)\.-]{3,})\s+([$]?\s*[0-9][0-9,]*\.\d{2})\s*$', ln)
        if m:
            label = m.group(1).strip()
            amount = float(re.sub(r'[^\d.]', '', m.group(2)))
            subtotal_candidates.append({'page':i, 'label':label, 'amount':amount, 'line':ln})
        # also capture lines with "Subtotal" or "Total"
        mm = re.search(r'\b(subtotal|sub total|sub-total|grand total|final total|total amount|amount due|net amount)\b.*?([0-9][0-9,]*\.\d{2})', ln, re.I)
        if mm:
            amt = float(re.sub(r'[^\d.]', '', mm.group(2)))
            final_total_candidates.append({'page':i, 'label':mm.group(1), 'amount':amt, 'line':ln})

# if no explicit "final total", try last page bottom-most numeric (heuristic)
if not final_total_candidates:
    # search for last numeric-looking amount in all pages
    last_amount = None
    for i, txt in all_page_texts:
        for ln in txt.splitlines()[::-1]:
            ln = ln.strip()
            if not ln: continue
            m = re.search(r'([0-9][0-9,]*\.\d{2})\s*$', ln)
            if m:
                last_amount = float(re.sub(r'[^\d.]', '', m.group(1)))
                final_total_candidates.append({'page':i, 'label':'inferred_last_amount', 'amount': last_amount, 'line': ln})
                break
        if final_total_candidates:
            break

# prepare result dict
result = {
    'master_items_count': len(master),
    'master_items': master,
    'sum_extracted_unique': round(sum_extracted, 2),
    'subtotal_candidates': subtotal_candidates,
    'final_total_candidates': final_total_candidates
}

# try to compute mismatch if OCR-reported final total exists
ocr_total = final_total_candidates[0]['amount'] if final_total_candidates else None
if ocr_total is not None:
    result['ocr_total'] = ocr_total
    result['difference'] = round(ocr_total - sum_extracted, 2)
    pct = (result['difference'] / ocr_total * 100) if ocr_total != 0 else None
    result['pct_difference'] = round(pct,2) if pct is not None else None

print(json.dumps(result, indent=2, ensure_ascii=False))

```

```python
# Cell: cleaning master items and robust total detection

import re, json

# Use master from previous step (variable 'master' or 'result["master_items"]')
try:
    prior_master = master
except NameError:
    prior_master = result.get('master_items', []) if 'result' in globals() else []

# Build cleaned master: remove clearly invalid items
clean_master = []
for it in prior_master:
    name = it.get('item_name','').strip()
    amt = it.get('item_amount') or 0.0
    page_no = it.get('_page_no', '')
    low = name.lower()

    # heuristics to treat as garbage / skip:
    if re.match(r'^[\d\.,\s]+$', name):          # name is basically numbers like "1000.00"
        continue
    if 'tem' in low or 'item quantity' in low or 'item amount' in low or 'tem_amount' in low:
        continue
    if page_no == '3':  # we saw page 3 is noisy  skip its items by default
        # keep only if name looks good and amount > 50 (optional)
        if amt < 100: 
            continue

    # otherwise accept
    clean_master.append(it)

# recompute sum
sum_clean = round(sum((it.get('item_amount') or 0.0) for it in clean_master), 2)

# search for explicit total lines across pages (more robust)
total_lines = []
for i, txt in enumerate(all_page_texts, start=1):
    page_idx, page_txt = txt if isinstance(txt, tuple) else (i, txt)
# NOTE: all_page_texts earlier was list of tuples (i, text). Use that:
total_lines = []
for (i, page_txt) in all_page_texts:
    for ln in page_txt.splitlines():
        ln2 = ln.strip()
        if not ln2: continue
        # match lines containing 'total' or 'amount due' etc with a numeric amount
        m = re.search(r'\b(grand total|final total|total amount|amount due|net payable|grand-total|total)\b.*?([0-9][0-9,]*\.\d{2})', ln2, re.I)
        if m:
            amt = float(re.sub(r'[^\d.]','', m.group(2)))
            total_lines.append({'page': i, 'line': ln2, 'amount': amt, 'match': m.group(1)})
        else:
            # also capture lines like "Total 1,23,456.00" (keyword then number)
            m2 = re.search(r'(total|amount due|net payable)[^\d]{0,10}([0-9][0-9,]*\.\d{2})', ln2, re.I)
            if m2:
                amt = float(re.sub(r'[^\d.]','', m2.group(2)))
                total_lines.append({'page': i, 'line': ln2, 'amount': amt, 'match': m2.group(1)})

# pick largest explicit total if any
ocr_reported_total = None
if total_lines:
    # prefer lines on later pages or with 'grand' in match, choose largest amount
    total_lines_sorted = sorted(total_lines, key=lambda x: (('grand' in x['match'].lower()) if x['match'] else False, x['amount']), reverse=True)
    ocr_reported_total = total_lines_sorted[0]['amount']

# fallback: sum of subtotal candidates if any
sum_subtotals = round(sum(s['amount'] for s in subtotal_candidates), 2) if 'subtotal_candidates' in globals() else None

# prepare reconciliation suggestions
suggested_final_total = None
method = None
if ocr_reported_total:
    suggested_final_total = ocr_reported_total
    method = 'ocr_reported_total'
else:
    # If sum_subtotals exists and is sensible, prefer it
    if sum_subtotals and sum_subtotals > 0:
        suggested_final_total = sum_subtotals
        method = 'sum_subtotals'
    else:
        suggested_final_total = sum_clean
        method = 'sum_clean_items'

reconciliation = {
    'clean_master_count': len(clean_master),
    'clean_master_sum': sum_clean,
    'ocr_reported_total': ocr_reported_total,
    'sum_of_subtotals': sum_subtotals,
    'suggested_final_total': suggested_final_total,
    'suggested_method': method
}

print(json.dumps({
    'clean_master_preview': clean_master[:8],
    'reconciliation': reconciliation
}, indent=2, ensure_ascii=False))
```
